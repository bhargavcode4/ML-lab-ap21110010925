```python
#Q4 Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
```

```python
import matplotlib.pyplot as plt
file_path = "C:\\Users\\Bhargav\\Desktop\\datas.txt"
x_label = 'Values on X-axis'
y_label = 'Values on Y-Axis'
title = 'Line Plot of the Values'
with open(file_path, 'r') as file:
    lines = file.readlines()
    data = [list(map(float, line.strip().split())) for line in lines]
x_values = [item[0] for item in data]
y_values = [item[1] for item in data]
plt.plot(x_values, y_values, label='Line Plot',color='r')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.legend()
plt.show()
```

```python
#Q7 Create plot usingsubplots(),scatter, and some other methods on theAxesobject
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x, y = np.random.rand(50), np.random.rand(50)
fig, axs = plt.subplots(2, 2, figsize=(6, 6))
axs[0, 0].scatter(x, y, color='blue', label='Scatter')

axs[0, 1].plot(x, y, color='green', linestyle='--', marker='o', label='Line')

bar_data = {'Cat-1': 20, 'Cat-2': 25, 'Cat-3': 30}
axs[1, 0].bar(bar_data.keys(), bar_data.values(), color='purple', label='Bar graph',edgecolor='black',width=0.5)

axs[1, 1].hist(np.random.randn(1000), bins=30, color='yellow', label='Histogram',edgecolor='black')
plt.show()
```

```python
#Q8 Generate random data for x and y and create a scatter plot using Matplotlib.
```

```python
np.random.seed(42)
num_pts = 1000
x_data = np.random.rand(num_pts)
y_data = 2 * x_data + 1 + 0.1 * np.random.randn(num_pts) 
plt.scatter(x_data, y_data, label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('points of Random Data')
plt.legend()
plt.show()
```

```python
#Q11 Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
```

```python
import matplotlib.pyplot as plt
categories = ['stripend', 'trading', 'freelance', 'youtube', 'writing']
expenses = [11000, 5000, 4500, 6000, 2550]
myexplode=[0.02,0.02,0.02,0.01,0.01]
plt.figure(figsize=(6, 6))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['teal', 'green', 'coral', 'yellow', 'pink'],explode=myexplode)
plt.title('Multiple incomes')
plt.show()
```
